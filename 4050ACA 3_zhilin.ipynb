```python
#Team member: Zhilin Zhai, Yining Liang

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
college= pd.read_csv("Downloads/CollegeScorecard.csv")

# we imported the data we need for this prject.
```

```python
college.head()
# show the data in the file so we can check if this is the correct data we are looking for.
```

```python
college2=college[['COSTT4_A','SATVRMID','PCTFLOAN','TUITIONFEE_OUT','TUITIONFEE_IN']]
# We chose five elements in "Collegescore card" to determine and analyze the information we want to know. 
#"COSTT4_A" is the average cost of attendance (academic year institutions) We 
#“SATVRMID” is the midpoint of SAT scores at the institution (critical reading). We 
#“PCTFLOAN” is the percent of all federal undergraduate students receiving a federal student loan
#"TUITIONONFEE_OUT" & "TUITIONFEE_IN" stand for the tuition that students(for 'outsate'&'instate') are going to pay. 
# We believe that all five elements listed above might affect the graduation rate for underpriveleged populations.
# For example, a lower tuition rate might be more acceptable and afforable for underpriveleged populations. A higher score of SAT might demonstate that this student have more money on getting a tutor for his or her study. Offering student loans might also help student of unprivileged population be able to afford their tuition and continue studying until they graduate. 
```

```python
college2
# data with our new-selected five elements
```

```python
Collegeclean = college2.dropna()
#clean the data - delete unuseful cells (empty), named it 'Collegeclean'.
```

```python
Collegeclean
#list out the 'Collegeclean' data and double check if it is the correct data we wanted.
```

```python
#####K means for cluster 2 
```

```python
from sklearn.cluster import KMeans
#import to help us simulate some data
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(Collegeclean)
# We first tried 2 clusters with the five elements we chose in 'Collegeclean'.
```

```python
kmeans.fit(Collegeclean)
```

```python
nd_Collegeclean=Collegeclean.values
nd_Collegeclean
#listed out data with five elements
```

```python
import matplotlib.pyplot as plt
```

```python
cluster_college=kmeans.predict(nd_Collegeclean)
```

```python
plt.scatter(nd_Collegeclean[:, 0], nd_Collegeclean[:, 1], c=cluster_college, cmap='viridis') # we used cluster_means as the indicator for clusters.
            
centers= kmeans.cluster_centers_

plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5); 
#'c' is for color, 's' is the dot size, 'alpha' is the tranparency.
```

```python
from sklearn.metrics import silhouette_score
# we then use silhouette plot to determine the number of clusters.
```

```python
silhouetteScore = silhouette_score(nd_Collegeclean, cluster_college, metric='euclidean')
```

```python
print('Silhouetter Score:', silhouetteScore)
```

```python
from yellowbrick.cluster import SilhouetteVisualizer
```

```python
silhouette_visualizer(kmeans, Collegeclean, colors='yellowbrick')
# we first tried 2 clusters.
# as a result, we got a silhouetter score of 0.5763 (rounding off 4 decimal places). 
#The score is 0.5763 and the silhouette plot seemed not too bad. 
#Since cluster 0 has a lot more similarities while cluster 1 is more heterogeneous. So we probably want to increase the cluster number.
```
