```python
df['ViolentCrime'] = df['ViolentCrime'].str.replace(',','')
df['PropertyCrime'] = df['PropertyCrime'].str.replace(',','')
df['Burglary'] = df['Burglary'].str.replace(',','')
df['Theft'] = df['Theft'].str.replace(',','') 
# we search the code from Google (zzl)
# we replace the replace the comma in the dataset to empty (this step only can use when the element are string) we have to redefine the string into float in the next step.
```

```python
df['ViolentCrime']=df['ViolentCrime'].astype(float)
df['PropertyCrime']=df['PropertyCrime'].astype(float)
df['Burglary']=df['Burglary'].astype(float)
df['Theft']=df['Theft'].astype(float)
# we search the code from Google (zzl)
# In order to calculate out the mean, we redefine the the element into float 
```

```python
crime_data=df
```

```python
crime_data.describe(include='all')
# We calculate out the mean of each elements (zzl)
```

```python
crime_data_mask =(crime_data['ViolentCrime']<= crime_data['ViolentCrime'].mean())&(crime_data['PropertyCrime'] <= crime_data['PropertyCrime'].mean())&(crime_data['Burglary'] <= crime_data['Burglary'].mean())&(crime_data['Theft'] <= crime_data['Theft'].mean())&(crime_data['MotorVehicleTheft'] <= crime_data['MotorVehicleTheft'].mean())
crime_data[crime_data_mask] 
# we filter out the cities which have the lower crime rate by comparing the factors of  "violent crime""property crime" "burglary""Theft""Motervehicle theft" with their mean.
# (zzl)
```

```python
import csv
f = open('college_data.csv')
csv_f_1 = csv.reader(f)
f = open('crime_data.csv')
csv_f_2 = csv.reader(f)

c_1 = [] #to define a variable for further use: to run for loop later
c_2 = []
for city_1 in csv_f_1: #in csv_f_1= in college data
    c_1.append(city_1[-1]) # only took last row in the data csv_f_1(which is college_data), which is "city".
for city_2 in csv_f_2:
    c_2.append(city_2[-1]) #only took last row in the data csv_f_2(which is crime_data), which is "city".

index_1 = [] #creat an empty list
index_2 = [] 
for i in range(len(c_1)): #for loop
    for j in range(len(c_2)): ##for loop 
        if c_1[i] == c_2[j]:      #condition: if c_1(city in college data) equals c_2(city in crime data)
                                  #for those cities which exist in both data, we collected them out and put it in an empty file for further use. 
            index_1.append(i)
            index_2.append(j)
print(index_1)
#print(index_2)  

#ynl,yrs
```

```python
vx = [] #created new empty list
simp = [] #created empty list,simp=simplified list

for i in range(len(c_1)): #for loop
    for j in range(len(c_2)):# for loop
        if c_1[i] == c_2[j]:#if data in c_1 equals the data in c_2, we put it into vx. for data that are not equal, it iterats over a sequence.
            vx.append(c_2[j])#at the end, this list will give us all the cities repeated in two data sets (college&crime)

for i in vx:# for loop,take every element in vx to run
    if i not in simp:#if any element in vs doesn't exist in simp, if any element is already in simp, it will be added again into simp
        simp.append(i)# will be added into simp.
print(simp)#print simp and check if it is the list we expected to have, clean data

#ynl
```

```python
lines = []#creat an empty list
with open('college_data.csv', 'r') as readFile: 

    reader = csv.reader(readFile)#read this file
    for row in reader:
        lines.append(row)#every column will be added into lines, which means lines=readerfile

    sample = []#emptey list
    
    for row in lines:#for loop, for every row in "lines"
        k = (row[-1])# read last element in the list, which is "city"
        if k in simp:
            sample.append(row) # if "k"（city) in "simp",we will put that row(for information,not just city name column anymore) in "sample"

with open('VXtest_1.csv','w') as writeFile:
    writer = csv.writer(writeFile)
    writer.writerows(sample)#write every row in sample/write this file
    
#ynl
```
