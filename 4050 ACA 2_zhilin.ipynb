```python
X_train =  ACAData[['GRADE', 'Gender', 'OBSNUM', 'totalobs-forsession', 'Activity', 'TRANSITIONS','NumACTIVITIES', 'FORMATchanges', 'NumFORMATS', 'Obsv/act','Transitions/Durations', 'Total Time']]
X_test =  ACAData2[['GRADE', 'Gender','OBSNUM', 'totalobs-forsession', 'Activity', 'TRANSITIONS','NumACTIVITIES', 'FORMATchanges', 'NumFORMATS', 'Obsv/act','Transitions/Durations', 'Total Time']]
Y_train = ACAData['ONTASK']
Y_test = ACAData2['ONTASK']

## # We decide not to use 'UNIQUEID', 'SCHOOL','CODER''Class' and 'STUDENTID' to train our model, because they are not explanatory variables. 
```

```python
X_train_dum = pd.get_dummies(X_train , columns=["Activity"], prefix=["Activity_Type"])
#Create Dummy variable for 'Activity'and transfer to multi-column binary --- from ICE 4
```

```python
X_test_dum = pd.get_dummies(X_test, columns=["Activity"], prefix=["Activity_Type"])
#Create Dummy variable for 'Activity'and transfer to multi-column binary----from ICE 4
```

```python
X_train_dum
```

```python
## Transform Y into binary variable 
Y_train.replace('Y',1,inplace=True)
Y_train.replace('N',0,inplace=True)
Y_test.replace('Y',1,inplace=True)
.replace('N',0,inplace=True)
```

```python
# Logicistic Regression ----Zhilin &Yining
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
# The classification report is a performance evaluation metric in machine learning. It can help us test the precision of training model.
# We search the classification report from Pandas library and google
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train_dum,Y_train)
```

```python
logmodel.intercept_
```

```python
logmodel.coef_
```

```python
Y_predict_train = logmodel.predict(X_train_dum)
```

```python
print(classification_report(Y_train, Y_predict_train))
# According to the classification report, the accuracy of our model is 0.67 
```

```python
# Decision tree ----Zhilin &Yining
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix
```

```python
NB_Gaussian = GaussianNB()
NB_Gaussian.fit(X_train_dum,Y_train)
Y_predict_train = NB_Gaussian.predict(X_train_dum)
print(classification_report(Y_train, Y_predict_train))
```

```python
### For Validation dataset
Y_predict_test = NB_Gaussian.predict(X_test_dum)
print(classification_report(Y_test, Y_predict_test))
# According to the classification report, the accuracy of our model is 0.62
```
